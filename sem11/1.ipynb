```python
from sympy import *
from math import *
```

```python
x = symbols('x')

y = 5*x**2 + 10*x - 30

```

```python
def disc(a,b,c):
    d = b**2 - 4*a*c
    x1 = (-b + sqrt(d))/(2*a)
    x2 = (-b - sqrt(d))/(2*a)
    return x1, x2
```

```python
print(disc(5,10,-30))
```

```python
print(solve(0<y))
```

```python
roots = solve(y, x)
roots
```
